```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, rfft, rfftfreq, fftfreq
import numpy as np

#open the document
var = pd.read_csv("930-data-export.csv")
#axis
x = list(var['Timestamp (Hour Ending)'])
y = list(var['Total CAL Demand (MWh)'])


#graph
plt.style.use('seaborn')
plt.plot(x,y)
plt.title("Excel sheet to Plot")
plt.xlabel("Fecha")
plt.ylabel("Demanda en MWh")
plt.show()

#format from dat to time
x2=var['Timestamp (Hour Ending)'] = pd.to_datetime(var['Timestamp (Hour Ending)']).dt.strftime("%H")

Q1 = len(y)



#
yf = abs(fft(y))[:Q1//2]
#fftfreq obtains the frequency od the data, retuns array of Q1 element,
#spaced by 3600 seconde
xf = fftfreq(Q1,3600)[:Q1//2]
#graph frequencies
plt.plot(xf, yf)
plt.grid()
plt.ylabel('Demand MWh')
plt.xlabel('frequency Hz')
plt.show()

#we want to graph in terms of time for better visualization so we go from the freqeucny domain to time domain.

t2=(1/xf)/3600
#graficamos
plt.plot(t2,yf)
plt.ylabel('Demand  MWh')
plt.xlabel('Hours')
plt.show()
```
